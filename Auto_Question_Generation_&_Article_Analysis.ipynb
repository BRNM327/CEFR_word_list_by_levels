## 讀入文章
*請上傳並使用純文字的.txt檔，再將第一行的Welcom Bear.txt改成新的檔案名稱即可*

```python
text = open("Welcom Bear.txt", "r", encoding="utf-8").read()
text = re.sub("\n", " ", text)
text_for_analysis = re.sub("[\,\"\'\.\?\!\:\;\r\n\”\’\?\“]", "", text)
text
```

## Article Analysis

```python
r = readability.Readability(text)
f = r.flesch()
print("Flesch Reading Ease score: {}".format(round(f.score, 1)))
print("Flesch Reading Ease readability: {}".format(f.ease))
print("Suitable stduent grade: {}".format(f.grade_levels[0]))
print(f"\n{'-'*35}")

doc = nlp(text_for_analysis.lower())
vocabs = list(set([token.lemma_ for token in doc]))

A1_words = []
A2_words = []
B1_words = []
B2_words = []
C1_words = []
C2_words = []

for v in vocabs:
    if v in A1:
        A1_words.append(v)
    if v in A2:
        A2_words.append(v)
    if v in B1:
        B1_words.append(v)
    if v in B2:
        B2_words.append(v)
    if v in C1:
        C1_words.append(v)
    if v in C2:
        C2_words.append(v)

print("Number of words in article: {}".format(len(doc)))
print("Number of vocabularies in article: {}".format(len(vocabs)))
print(f"\n{'-'*35}")
print(f"""Percentage of different levels:
A1: {round(len(A1_words)/len(vocabs)*100, 2)}%, {len(A1_words)} vocabularies
vocabularies: {", ".join(list(set(A1_words)))}

A2: {round(len(A2_words)/len(vocabs)*100, 2)}%, {len(A2_words)} vocabularies
vocabularies: {", ".join(list(set(A2_words)))}

B1: {round(len(B1_words)/len(vocabs)*100, 2)}%, {len(B1_words)} vocabularies
vocabularies: {", ".join(list(set(B1_words)))}

B2: {round(len(B2_words)/len(vocabs)*100, 2)}%, {len(B2_words)} vocabularies
vocabularies: {", ".join(list(set(B2_words)))}

C1: {round(len(C1_words)/len(vocabs)*100, 2)}%, {len(C1_words)} vocabularies
vocabularies: {", ".join(list(set(C1_words)))}

C2: {round(len(C2_words)/len(vocabs)*100, 2)}%, {len(C2_words)} vocabularies
vocabularies: {", ".join(list(set(C2_words)))}""")

clean_vocabs = [token.lemma_ for token in doc if token.lemma_ not in stopwords]
counts = Counter(clean_vocabs)
freq_dict = sorted(dict(counts).items(), key=lambda x: x[1], reverse=True)

print(f"\n{'-'*35}")
print("Top ten most frequent words (Word | Frequency):")
for i in range(10):
    print(freq_dict[i][0]+": "+str(freq_dict[i][1]))

doc2 = nlp(text.lower())

keyness = {}
for phrase in doc2._.phrases:
    keyness[phrase.text] = phrase.rank

keyness_sorted = sorted(keyness.items(), key=lambda x: x[1], reverse=True)

print(f"\n{'-'*35}")
print("Top ten most important words  (Word | Importance):")
for i in range(10):
    print(keyness_sorted[i][0]+": "+str(round(keyness_sorted[i][1]*100, 2)))

```

## Question Generation

### Yes/No Questions

```python
payload = {"input_text": text}
boutput = qe.predict_boolq(payload)

q = boutput.get('Boolean Questions')[0]
payloadfa = {"input_text" : text,
        "input_question" : q}

banswer = answer.predict_answer(payloadfa)
print(f"\n{'-'*35}")
print("Question: "+q+"\n"+"Answer: "+banswer)
```

### MCQ Questions
*第2行的max_questions後面的數字可以選擇要出幾題*

```python
payload = {"input_text": text, "max_questions": 5}
moutput = qg.predict_mcq(payload)

print(f"\n{'-'*35}")
for i, mcq in enumerate(moutput.get("questions")):
    print(str(i+1)+". "+mcq.get("question_statement"))
    print("\u2705 "+mcq.get('answer'))
    for o in mcq.get('options'):
        print("\u274C "+o.lower())
    for eo in mcq.get('extra_options'):
        print("\u274C "+eo.lower())
    print()
```

### FAQ Questions
*第2行的max_questions後面的數字可以選擇要出幾題*

```python
payload = {"input_text": text, "max_questions": 5}
foutput = qg.predict_shortq(payload)

print(f"\n{'-'*35}")
for i, faq in enumerate(foutput.get("questions")):
    print(str(i+1)+". "+faq.get('Question')+"\n", "Answer: "+faq.get('Answer'))
```
